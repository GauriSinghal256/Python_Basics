## Advance operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restraurant_id , 2021 , 2022 , 2023 , 2024]
sales_data = np.array([
    [1 , 150000, 180000 , 220000 , 250000],  # Paradise Biryani
    [2 , 120000, 140000 , 160000 , 190000],  # Beijing Bites
    [3 , 200000, 230000 , 260000 , 300000],  # Pizza Hub
    [4 , 180000, 210000 , 240000 , 270000],  # Burger Point
    [5 , 160000, 185000 , 205000 , 230000],  # Chai Point   
])


print("==== Zomato sales analysis ====")
print("\n Sales data shape" , sales_data.shape)
print("Sales data for first 3 restraunts \n " , sales_data[:3])
print('\n')
print( sales_data[: , 1:])
```

```python
# Total sales per year
print(np.sum(sales_data , axis=0))
yearly_total = np.sum(sales_data[: , 1:] , axis=0)
print(yearly_total)

# Minimum sales per restaurant 
min_sales = np.min(sales_data[: , 1:] , axis =1)
print(min_sales)

# Maximum Sales per year 
max_sales = np.max(sales_data[: , 1:] , axis=0)
print(max_sales)

# Average sales per restraunt 
avg_sales = np.mean(sales_data[: , 1:] , axis = 1)
print(avg_sales)

# Cummulative sales 
cumSum = np.cumsum(sales_data[: , 1:] , axis=1)
print(cumSum)

plt.figure(figsize=(10 ,6))
plt.plot(np.mean(cumSum , axis=0))
plt.title("Average Cumulative sales across all restraunts")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,5])
print("Vector addition" , vector1+vector2)
print("Multiplication vector" , vector1*vector2)
print("Dot product" , np.dot(vector1 , vector2))
angle = np.arccos(np.dot(vector1 , vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```

```python
# basically here if we want to perform some task on each and every element of the array we use vectorize
restaurant_types = np.array(['biryani' , 'chinese' , 'pizza' , 'burger' ,'cafe'])
vectorized_upper = np.vectorize(str.upper)
print(vectorized_upper(restaurant_types))
```

```python
# Here when we divide by 12 (scaler) so we can directly perform it we wont need to convert it to vectorized form
# In scalar this broadcasting is done. It automatically go to every part of the array and perform the operation
monthly_avg = sales_data[: , 1:]/12
print(monthly_avg)
```
