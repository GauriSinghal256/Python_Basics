```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices = [60,40]
plt.pie(slices)

plt.title("My awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices = [60,40]
labels = ['Sixty' , 'Forty']
plt.pie(slices ,labels=labels, wedgeprops={'edgecolor':'black'})

plt.title("My awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("fivethirtyeight")

slices = [120,80,30,40]
labels = ['Sixty' , 'Forty' , 'Extra1' , 'Extra2']
colors = ['blue' , 'red' , 'yellow' , 'green']
plt.pie(slices ,labels=labels,colors=colors ,wedgeprops={'edgecolor':'black'})

plt.title("My awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

slices = [59219 , 55466,47544,36443,35917,31191 , 27097,
          23030 , 20524, 18523, 18017, 7920 , 7331 , 7201, 5833]
labels = ['Javascript' , 'HTML/CSS' , 'SQL' , 'Python' , 'Java' , 'Bash/Shell/Powershell' , 
          'C#' , 'PHP' , 'C++' , 'TypeScript' , 'C' , 'Other(s):' , "Ruby" , 'Go' , 'Assembly']

plt.pie(slices , labels=labels ,wedgeprops={'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

slices = [59219 , 55466,47544,36443,35917]
labels = ['Javascript' , 'HTML/CSS' , 'SQL' , 'Python' , 'Java']
explode =[0,0,0,0.1,0]

plt.pie(slices , labels=labels, explode=explode , shadow=explode, startangle=90 ,wedgeprops={'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# 0.1 means that 10 percent of the radius 
# shadow = true for shadow 
# if here we want to rotate it then we can do startangle
```

```python
# Display the perentage 
plt.style.use('fivethirtyeight')

slices = [59219 , 55466,47544,36443,35917]
labels = ['Javascript' , 'HTML/CSS' , 'SQL' , 'Python' , 'Java']
explode =[0,0,0,0.1,0]

plt.pie(slices , labels=labels, explode=explode , shadow=explode, startangle=90, autopct='%1.1f%%' ,wedgeprops={'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

```
